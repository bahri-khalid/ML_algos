```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../resources/univariate_linear_regression_dataset.csv')
data.columns = ["X","y"]
x = data.X.values
X = np.vstack((x**2,x,np.ones(x.shape))).T
y = data.y.values
y.shape = (len(y),1)
```

```python
plt.grid()
plt.scatter(x,y)
plt.show()
```

```python
# theta shoud be represented by a columns
def model(X,theta):
    return X.dot(theta)
```

```python
theta = np.random.randn(3,1)
predicted = model(X,theta)
plt.grid()
plt.scatter(x,y)
plt.scatter(x,predicted,c='orange')
plt.show()
```

```python
def loss(predicted,y):
    m = len(y)
    return np.sum((predicted-y)**2)/(2*m)
loss(predicted,y)
```

```python
def gradient(X,y,theta):
    m = len(y)
    return (X.T.dot(X.dot(theta)-y))/m
```

```python
def gradientDescent(X,y,theta,learning_rate = 0.01,n_iterations = 100):
    th = theta.copy()
    for i in range(n_iterations):
        th -= learning_rate*gradient(X,y,th)
    return th
newTheta = gradientDescent(X,y,theta)
newPrediction = model(X,newTheta)
print('mean square error = ',loss(newPrediction,y))
plt.grid()
plt.scatter(x,y)
plt.scatter(x,newPrediction,c = 'red')
plt.show()
newTheta
loss(newPrediction,newY)
```
