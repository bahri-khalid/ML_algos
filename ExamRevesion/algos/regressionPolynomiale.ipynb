```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../resources/dataset_polynomial3.csv")
data.head()
x = data.x
y = data.y
plt.grid()
plt.scatter(x,y,c = 'red')
plt.show()
```

```python
dgp = 3
p = np.poly1d(np.polyfit(x,y,dgp))
plt.grid()
LPol = list(p)
x1 = np.linspace(max(x),min(x),100)
plt.plot(x1,p(x1),c = 'red')
plt.scatter(x,y,c = 'green')
plt.show()
```
