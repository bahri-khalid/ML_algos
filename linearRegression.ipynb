```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Load the data from the diabetes dataset
# Use the return_X_y = True to split the dataset into features/target
X,y = datasets.load_diabetes(return_X_y = True)
# the X contain 10 features we use the newaxis to split axis and take only one, the first for example
X = X[:,np.newaxis,6]

# Now we are going to split data in to train and test datasets
X_train = X[:-100]
X_test = X[-100:]

y_train = y[:-100]
y_test = y[-100:]

# institiat the linear regression object
model = LinearRegression()

# make the training "feed the model the data"
 
model.fit(X,y)

# now use the test data to validate our model
print('MSR = ',mean_squared_error(y_test,model.predict(X_test)) )
print('R2  = ',r2_score(y_test,model.predict(X_test)) )

plt.scatter(X_test,y_test)
plt.scatter(X_test,model.predict(X_test))
plt.show()
print(model.coef_)
print(model.intercept_)

d = X_test * model.coef_ + model.intercept_
plt.scatter(X_test,d)
```
